Libraries

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
```

```python
# Features: Alter, Sportlich, Mag Fast Food, Budget pro Besuch beim Restaurant
# Prediction: Mag Patties N' Berries

alter = np.array([16, 18, 18, 18, 19, 21, 40, 9, 10, 16, 32, 43, 70, 63, 25, 27, 28])
sportlichkeit = np.array([7.5, 8.5, 6.5, 5.0, 8.0, 10, 4.5, 7.0, 6.5, 5.3, 8.0, 7.9, 2.0, 7.1, 4.9, 6.4, 8.9])
X = np.c_[alter, sportlichkeit]
y = np.array([1, 1, 1, 1, 0, 0, 1, 1, 1, 1, 0, 1, 0, 0, 1, 1, 1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train)
print("\n")
print(X_test)

plt.scatter(X_train[y_train==0][:, 0], X_train[y_train==0][:, 1], c="red", label="Mag nicht Patties N' Berries - Train")
plt.scatter(X_train[y_train==1][:, 0], X_train[y_train==1][:, 1], c="green", label="Mag Patties N' Berries - Train")
plt.scatter(X_test[y_test==0][:, 0], X_test[y_test==0][:, 1], c="red", label="Mag nicht Patties N' Berries - Test", marker="x")
plt.scatter(X_test[y_test==1][:, 0], X_test[y_test==1][:, 1], c="green", label="Mag Patties N' Berries - Test", marker="x")
plt.xlabel("Alter")
plt.ylabel("Sportlichkeit")
plt.show()
```

Sklearn Classifier Lib's

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train, y_train)

plot_tree(
    tree_classifier,
    feature_names=["Alter", "Sportlichkeit"],
    class_names=["Mag nicht Patties N' Berries", "Mag Patties N' Berries"],
    rounded = True,
    filled = True
)

print(tree_classifier.score(X_test, y_test))

plt.show()
```
