```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

depression_dataset = pd.read_csv("../Datasets/depression.csv", delimiter=",")

suicidal_thoughts = depression_dataset[depression_dataset["Have you ever had suicidal thoughts ?"] == "Yes"]
print(len(suicidal_thoughts), f"students out of {len(depression_dataset)} have suicidal thoughts")

print("First 10 students:\n")
rechteckiger_ausschnitt = depression_dataset.iloc[0:11, depression_dataset.columns.get_indexer(["Academic Pressure", "Sleep Duration", "Have you ever had suicidal thoughts ?"])]
print(rechteckiger_ausschnitt)
print("\n")
cgpa = depression_dataset["CGPA"][0:11].astype("int")
print("CGPA's converted to ints: ", list(cgpa))
print("\n")

ages = sorted(set(list(depression_dataset["Age"])))
print(ages)
average_academic_pressure = list(depression_dataset.groupby("Age")["Academic Pressure"].mean())
print(average_academic_pressure)

fig, ax = plt.subplots()

ax.set_xlabel("Age")
ax.set_ylabel("Academic Pressure")
ax.plot(ages, average_academic_pressure)

plt.show()
```
