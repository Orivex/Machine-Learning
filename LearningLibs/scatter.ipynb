```python
import numpy as np
import matplotlib.pyplot as plt

# Zufällige Werte für 10 Personen
anzahl_personen = 10
gespielteSpiele = np.random.rand(anzahl_personen) * 30
gespielteStunden = np.random.rand(anzahl_personen) * 1000 
iq = np.random.rand(anzahl_personen) * 200

# Scatterplot
fig, ax = plt.subplots()
ax.scatter(gespielteSpiele, gespielteStunden, s=iq)

# Achsentitel
ax.set_xlabel("Gespielte Spiele")
ax.set_ylabel("Gespielte Stunden")
ax.set_title("Beziehung zwischen Spielen, Stunden und IQ")

# Grafik anzeigen
plt.show()
```
